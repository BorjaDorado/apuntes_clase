# 2.2- Visualización con Pandas

https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html


```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd
import numpy as np

import pylab as plt
import seaborn as sns
```

**Line plot**

```python
df=pd.Series(np.random.randn(1000),
             index=pd.date_range('1/1/2000', periods=1000))

df.cumsum().plot();
```

```python
df=pd.DataFrame(np.random.randn(1000, 4), 
                index=df.index, columns=list('ABCD'))


display(df.head())

df=df.cumsum()

df.plot();
```

```python
df.plot(x='A', y='B');
```

```python
df.plot.scatter(x='A', y='B');
```

**Bar plot**

```python
df.iloc[5].plot(kind='bar');
```

```python
df.iloc[5].plot(kind='bar')

plt.axhline(1, color='r')

plt.show();
```

```python
df=pd.DataFrame(np.random.rand(10, 4), 
                columns=['a', 'b', 'c', 'd'])


df.plot.bar();
```

```python
df.plot.bar(stacked=True);
```

```python
df.plot.barh(stacked=True);
```

**Histograma**

```python
df=pd.DataFrame({'a': np.random.randn(1000) + 1,
                 'b': np.random.randn(1000),
                 'c': np.random.randn(1000) - 1})
    
df.plot.hist();
```

```python
df[['c', 'b', 'a']].plot.hist(bins=30);
```

```python
df.plot.hist(bins=30, stacked=True);
```

```python
df['a'].plot.hist();
```

```python
df['a'].plot.hist(orientation='horizontal');
```

```python
help(df['a'].plot.hist)
```

**Box plots**

```python
df=pd.DataFrame(np.random.rand(10, 5), 
                columns=['A', 'B', 'C', 'D', 'E'])


df.plot.box();
```

```python
df.boxplot();
```

```python
color={'boxes': 'DarkGreen',
       'whiskers': 'DarkOrange',
       'medians': 'DarkBlue',
       'caps': 'Gray'}


df.plot.box(color=color, vert=False);
```

**Area plot**

```python
df.plot.area();
```

```python
df.plot.area(stacked=False);
```

**Scatter plot**

```python
df.plot.scatter(x='A', y='B');
```

```python
ax = df.plot.scatter(x='A', y='B')

df.plot.scatter(x='A', y='C', ax=ax, color='DarkGreen')

df.plot.scatter(x='B', y='C', ax=ax, color='DarkBlue');
```

```python
 df.plot.scatter(x='C', y='B', c='A');
```

```python
# esto tiene 4 dimensiones (x, y, color, tamaño)

df.plot.scatter(x='C', y='B', c='A', s=df['D']*100);
```

**Hexagonal bin plot**

```python
df=pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])

df['b']=df['b']+np.arange(1000)

df.plot.hexbin(x='a', y='b', gridsize=25);
```

**Pie plot**

```python
serie=pd.Series(3 * np.random.rand(4), 
                index=['a', 'b', 'c', 'd'], 
                name='serie')

serie.plot.pie(figsize=(6, 6));
```

```python
plt.rcParams["font.family"] = "serif"


serie.plot.pie(labels=['AA', 'BB', 'CC', 'DD'],
               colors=['r', 'g', 'b', 'c'],
               autopct='%.2f',
               fontsize=20,
               figsize=(6, 6));
```

```python
df=pd.DataFrame(3*np.random.rand(4, 2), 
                index=['a', 'b', 'c', 'd'], 
                columns=['x', 'y'])


df.plot.pie(subplots=True, figsize=(8, 4));
```

**Scatter matrix plot**

```python
pd.plotting.scatter_matrix(df);
```

**curvas de Andrews**
